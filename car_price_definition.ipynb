## Подготовка данных к обучению

### Разделение данных на выборки

Делим данные на обучающую (60%), валидационную (20%) и тестовую (20%) выборки.

```python
target = df_work['Price']
features = df_work.drop('Price', axis=1)

# Разделение на обучающую и валидационную+тестовую
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.4, random_state=12345)

# Разделение на валидационную и тестовую
features_valid, features_test, target_valid, target_test = train_test_split(
    features_test, target_test, test_size=0.5, random_state=12345)


print('features_train', features_train.shape, '\ntarget_train', target_train.shape, 
      '\nfeatures_valid', features_valid.shape, '\ntarget_valid', target_valid.shape,
      '\nfeatures_test', features_test.shape, '\ntarget_test', target_test.shape) # Проверка
```

### Кодирование категориальных признаков методом OHE

```python
# Список категориальных признаков
cat_features = ['VehicleType', 'Gearbox', 'Model', 'FuelType', 'Brand']
```

```python
# Кодирование категориальных признаков
ohe = OneHotEncoder(sparse=False, handle_unknown="ignore")

# Кодируем признаки обучающей выборки
x_ohe = ohe.fit_transform(features_train[cat_features])

# Добавляем новые закодированные признаки к численным признакам обучающей выборки
x_train = (features_train.drop(cat_features, axis=1).
                         join(pd.DataFrame(x_ohe, 
                                           columns=ohe.get_feature_names(cat_features), 
                                           index=features_train.index))
         )

# Кодируем признаки валидационной выборки
x_ohe = ohe.transform(features_valid[cat_features])

# Добавляем новые закодированные признаки к численным признакам тестовой выборки
x_valid = (features_valid.drop(cat_features, axis=1).
                        join(pd.DataFrame(x_ohe, 
                                          columns=ohe.get_feature_names(cat_features),
                                          index=features_valid.index))
         )

# Кодируем признаки тестовой выборки
x_ohe = ohe.transform(features_test[cat_features])

# Добавляем новые закодированные признаки к численным признакам тестовой выборки
x_test = (features_test.drop(cat_features, axis=1).
                        join(pd.DataFrame(x_ohe, 
                                          columns=ohe.get_feature_names(cat_features),
                                          index=features_test.index))
         )
print('Размерность обучающей выборки:', x_train.shape)
print('Размерность валидационнной выборки:', x_valid.shape)
print('Размерность тестовой выборки:', x_test.shape)
```

```python
# Меняем названия и проверяем:
features_train = x_train
display(features_train.head())
features_valid = x_valid
display(features_valid.head())
features_test = x_test
display(features_test.head())
```

### Масштабирование данных

```python
# Функция для масштабирования данных.
# Принимает на вход признаки обучающей, валидационной и тестовой выборок.
# Обучается на выборке train
# Привеняется ко всем

def scaler(features_train, features_valid, features_test):
    # Список всех численных признаков
    numeric = ['RegistrationYear', 'Power', 'Kilometer', 'RegistrationMonth', 'PostalCode']

    # Создаем объект структуры StandardScaler() и настраиваем его на обучающих данных 
    scaler = StandardScaler()
    scaler.fit(features_train[numeric])

    features_train[numeric] = scaler.transform(features_train[numeric])
    features_valid[numeric] = scaler.transform(features_valid[numeric])
    features_test[numeric] = scaler.transform(features_test[numeric])
    
    display(features_train[numeric].head()) # Проверка
    return features_train, features_valid, features_test
```

```python
# Проверка масштабирования
features_train, features_valid, features_test = scaler(features_train, features_valid, features_test)
```
