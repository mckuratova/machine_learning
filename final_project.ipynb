# Телеком — задача проекта

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

# Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

# Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

#### Преобразование данных

Преобразуем столбцы *BeginDate* и *EndDate* к формату *datetime*. Для столбца *EndDate* вместо значений No будем использовать заданную дату 01 февраля 2020 года, записанную в константу *NOW_DATE*. Полученные значения сохраним в новые столбцы *begin_date_dt и end_date_dt*.

*BeginDate*

```python
# Преобразуем столбец BeginDate в формат DateTime и проверяем результат
contract['begin_date_dt'] = pd.to_datetime(contract['BeginDate'], format='%Y-%m-%d')
contract.begin_date_dt.head()
```

*EndDate*

```python
# Преобразуем столбец EndDate в формат DateTime
# Столбец содержит значения No, заменим их на NaT
contract['end_date_dt'] = pd.to_datetime(contract['EndDate'], format='%Y-%m-%d', errors='coerce')
contract.end_date_dt.head()
```

```python
# Заполним пропущенные значения датой 1 февраля 2020 года
print('Количество пропусков в столбце end_date_dt:', contract.end_date_dt.isna().sum())
contract.end_date_dt.fillna(pd.Timestamp(NOW_DATE), inplace=True)
print('Количество пропусков в столбце end_date_dt (проверка):', contract.end_date_dt.isna().sum())
contract.end_date_dt.head()
```

*TotalCharges*

### Кодирование признаков

```python
# Список категориальных признаков
cat_features = ['Type', 'PaperlessBilling', 'PaymentMethod', 'InternetService']
```

```python
# Кодирование категориальных признаков
ohe = OneHotEncoder(sparse=False, drop='first')

# Кодируем признаки обучающей выборки
X_ohe = ohe.fit_transform(features_train[cat_features])

# Добавляем новые закодированные признаки к численным признакам обучающей выборки
X_train = (features_train.drop(cat_features, axis=1)
                         .join(pd.DataFrame(X_ohe, 
                                           columns=ohe.get_feature_names(cat_features), 
                                           index=features_train.index))
         )

# Кодируем признаки тестовой выборки
X_ohe = ohe.transform(features_test[cat_features])

# Добавляем новые закодированные признаки к численным признакам тестовой выборки
X_test = (features_test.drop(cat_features, axis=1).
                        join(pd.DataFrame(X_ohe, 
                                          columns=ohe.get_feature_names(cat_features),
                                          index=features_test.index))
         )
print('Размерность обучающей выборки:', X_train.shape)
print('Размерность тестовой выборки:', X_test.shape)
```

### Масштабирование данных

```python
def scaler(features_train, features_test):
    # Список всех численных признаков
    numeric = ['MonthlyCharges', 'TotalCharges', 'number_of_services', 'SeniorCitizen',
               'alone', 'MultipleLines', 'interval']

    # Создаем объект структуры StandardScaler() и настраиваем его на обучающих данных 
    scaler = StandardScaler()
    scaler.fit(features_train[numeric])

    features_train[numeric] = scaler.transform(features_train[numeric])
    features_test[numeric] = scaler.transform(features_test[numeric])
    
    display(features_train[numeric].head()) # Проверка
    return features_train, features_test
```

```python
scaled_X_train, scaled_X_test = scaler(X_train, X_test)
```

Модель CatBoost показала лучшее значение AUC. Проверим ее на тестовой выборке.

### Проверка лучшей модели на тестовой выборке

```python
# Функция строит ROC-кривую по целевым значениям и вероятностям класса "1",
# а также ROC-кривую случайной модели
def show_roc(target, probabilities):
    probabilities_one = probabilities[:, 1]
    fpr, tpr, thresholds = roc_curve(target, probabilities_one)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='green')

    # ROC-кривая случайной модели
    plt.plot([0, 1], [0, 1], linestyle='--', color='red')

    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC-кривая")
    plt.show()
    
    auc_roc = roc_auc_score(target, probabilities_one)
    print('AUC-ROC', auc_roc.round(4))
    return auc_roc
```

```python
test_cat_auc = show_roc(y_test, cat_probabilities)
```
